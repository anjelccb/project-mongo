```python
def geopoint(x): 
    for dic in x:
        longitude = dic["longitude"]
        latitude = dic["latitude"]
        
    
        return {"type": "Point", "coordinates": [longitude, latitude]}

df_companies['geo']= df_companies['offices'].apply(geopoint)
```
